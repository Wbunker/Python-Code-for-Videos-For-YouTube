# Let's see what it does by default

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

fontP = FontProperties()
fontP.set_size('xx-small')

p1, = plt.plot([1, 2, 3], label='Interest Rates')
p2, = plt.plot([3, 2, 1], label='Stock Prices')
plt.legend(handles=[p1, p2], title='Relationship', prop=fontP)
plt.show()
```

# 2 Main Properties Control This

## 1. loc - sets the legend's basic location

```python
## 'upper left', 'upper right', 'lower left', 'lower right'
## 'upper center', 'lower center', 'center left', 'center right'
## 'center'
## 'best'
```

```python
p1, = plt.plot([1, 2, 3], label='Interest Rates')
p2, = plt.plot([3, 2, 1], label='Stock Prices')
plt.legend(handles=[p1, p2], title='Relationship',loc='upper center',  prop=fontP)
plt.show()
```

## 2. bbox_to_anchor - allows you to offset this

```python
p1, = plt.plot([1, 2, 3], label='Interest Rates')
p2, = plt.plot([3, 2, 1], label='Stock Prices')
plt.legend(handles=[p1, p2], title='Relationship', bbox_to_anchor=(0, 0),  prop=fontP)
plt.show()
```

```python
p1, = plt.plot([1, 2, 3], label='Interest Rates')
p2, = plt.plot([3, 2, 1], label='Stock Prices')
plt.legend(handles=[p1, p2], title='Relationship', bbox_to_anchor=(1, 0),  prop=fontP)
plt.show()
```
