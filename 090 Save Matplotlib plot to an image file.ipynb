# Use savefig, and if file ending is recognized it will save correctly


### First we will make some data and show on screen

```python
import random
import matplotlib.pyplot as plt

random_list = [random.randint(0,50) for i in range(500)]


plt.plot(random_list)
plt.show()
```

### Now how to save to file

```python
# show clears the buffer so we need to plot again

plt.plot(random_list)

plt.savefig('data/random_plot.png')

plt.savefig('data/random_plot.pdf')
```
