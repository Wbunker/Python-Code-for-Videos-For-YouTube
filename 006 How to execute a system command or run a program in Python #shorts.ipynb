```python
import subprocess

results = subprocess.run(
        ['ls', '-l'], 
        text=True)
print(results.stdout)
```

```python
import time

results = subprocess.Popen(
            ['sleep', '6'], 
            text=True)

while results.poll() == None:
    # do some other work
    time.sleep(1)
    print("second of work")
    
stdout, stderr = results.communicate()


print("I have finished the command")
```
