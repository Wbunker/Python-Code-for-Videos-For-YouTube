```python
import requests
from time import sleep
import random
from concurrent.futures import ThreadPoolExecutor

def io_task(datacall):
    sleep(random.randint(1,1000)/1000)
    return (datacall, random.randint(1,10))

def handle_finished_task(future):
    print(future.result())
    
def handle_finished_other(future):
    print(f"Doing another task with {future.result()}")
```

```python
with ThreadPoolExecutor(max_workers=20) as pool:
    
    futures = [pool.submit(io_task, datacall) for datacall in range(40)]
    for future in futures:
        future.add_done_callback(handle_finished_task)
```

```python
with ThreadPoolExecutor(max_workers=20) as pool:
    
    futures = [pool.submit(io_task, datacall) for datacall in range(40)]
    for future in futures:
        future.add_done_callback(handle_finished_task)
        future.add_done_callback(handle_finished_other)
```
