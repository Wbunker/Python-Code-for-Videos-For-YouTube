# Tricky because you have to handle a lot of formats

## We will start with isdigit

```python
teststrings = ["1", "2.5", "264xdsrf", "023", '1e3']


for test in teststrings:
    print(f"{test}: {test.isdigit()}")
```

```python
for test in teststrings:
    cleaned_string = test.replace('.', '', 1)
    print(f"{test}: {cleaned_string.isnumeric()}")
```

# Use the EAFP principle



```python
def is_numeric(string_to_test):
    
    try:
        float(string_to_test)
        return True
    except ValueError:
        return False
    
    
for test in teststrings:
    print(f"{test}: {is_numeric(test)}")
```
